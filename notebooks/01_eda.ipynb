```python
import requests
import pandas as pd

# 1. Alamat API untuk mengambil closed issues dari proyek facebook/react
url = "https://api.github.com/repos/facebook/react/issues"
params = {
    "state": "closed",
    "per_page": 100
}

print("Sedang mengambil data dari GitHub API...")
response = requests.get(url, params=params)

if response.status_code == 200:
    issues_data = response.json()
    
    parsed_issues = []
    for issue in issues_data:
        if "pull_request" not in issue: 
            parsed_issues.append({
                "id": issue["id"],
                "number": issue["number"],
                "title": issue["title"],
                "state": issue["state"],
                "created_at": issue["created_at"],
                "closed_at": issue["closed_at"]
            })
            
    df = pd.DataFrame(parsed_issues)
    print(f"Berhasil mengambil {len(df)} data issue!")
    
    df.to_csv("../data/raw/raw_issues.csv", index=False)
    print("Data mentah berhasil disimpan di folder data/raw/raw_issues.csv")
else:
    print(f"Error Code: {response.status_code}")
    print("Gagal mengambil data, periksa koneksi internet Anda.")
```
